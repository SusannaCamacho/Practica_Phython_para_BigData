```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']
```

```python
# A. Una función que calcule el precio total de la compra

def total_compra(articulos, compra):
    total = 0
    for index in compra:
        total += articulos[index]
    return print(f'El precio total de la compra es: {total}€')


total_compra(articulos, compra)
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%

def total_compra(articulos, compra):
    total = 0
    for index in compra:
        total += articulos[index]
    descuento10 = total * 0.9
    return print(f'El precio total de la compra con el descuento del 10% es: {descuento10}€')

total_compra(articulos, compra)
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

def total_compra(articulos, compra):
    total = 0
    for index in compra:
        total += articulos[index]
    if total > 100:
        descuento10 = total * 0.9
    else:
        descuento10 = total
    return print(f'El precio total de la compra con el descuento del 10%, si supera los 100€ es: {descuento10}€')

total_compra(articulos, compra)
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%)

def total_compra(articulos, compra):
    total = 0
    for index in compra:
        total += articulos[index]
    iva21 = round(total * 1.21, 2)
    return print(f'El precio total de la compra con el 21% de IVA es: {iva21}€')

total_compra(articulos, compra)
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca el nombre, el apellido y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]
```

```python
def prom_notas_curso(curso, alumnos):
    lista_alumnos = []
    for index in range(len(alumnos)):
        if alumnos[index][2] == curso:
            nombre_apellido = alumnos[index][0] +" "+ alumnos[index][1]
            promedio = sum(alumnos[index][3]) / len(alumnos[index][3])
            lista_alumnos.append((nombre_apellido, promedio)) 
    return lista_alumnos
```
